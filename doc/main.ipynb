### 2.2 Power Dissipation Index (PDI)
PDI is a metric combining a storm's wind velocity and duration, in order to approximate the overall intensity of the hurricane.

```python
#@title Defining function to calculate PDI
def nan_helper(y):
  return np.isnan(y), lambda z: z.nonzero()[0]

def calculate_power_dissipation_index(storm):
  # INPUT: 
  # storm: a Xarray DATASET object (this can be adjusted if desired)
  # OUTPUT:
  # pdi

  storm_times = storm.time.values[~np.isnan(storm.time.values)]
  wind_speeds = storm.wmo_wind.values[~np.isnan(storm.time.values)] * 0.514444 # converting to m/s

  if (np.size(wind_speeds[~np.isnan(wind_speeds)])==0):
    pdi = np.nan
    return pdi
  
  nans,x = nan_helper(wind_speeds)
  wind_speeds[nans] = np.interp(x(nans),x(~nans),wind_speeds[~nans])
  pdi=0

  num_time_steps = np.size(storm_times)
  for i in np.arange(num_time_steps):
    V3 = wind_speeds[i] ** 3
    if(i==0):
      delta_t = 3 * 60 * 60 # standard time step, 3 hours (converted to seconds)
    else:
      delta_t = (storm_times[i] - storm_times[i-1]) * 24 * 60 * 60
    product = V3 * delta_t
    pdi = pdi + product

  pdi = pdi * 10**-11
  return pdi
```

```python
pdi = np.array([])
num_storms = cost.dims['storm'] 
for i in np.arange(num_storms):
  storm = cost.sel(storm=i)
  storm_pdi = calculate_power_dissipation_index(storm)
  pdi = np.append(pdi, storm_pdi)
```

```python
cost=cost.assign(pdi = (['storm'],pdi))
cost.pdi.attrs={"long_name": "power density index","Description": "Calculated as integral V^3 dt, as in Nakumara et al."}
```

### 2.3 Sea Surface Temperature (SST)

SST is also an important factor in understanding hurricanes, as warmer waters mean there is more potential energy for the storm to convert into motion. Therefore, we used two NOAA datasets containing gridded SST data at a monthly temporal resolution, and a 1 deg. spatial resolution. For 1961-1981, we only have climatological SST data (i.e. the monthly average for a given point over a several year span), but for 1981-now, we have monthly averages for each year. Therefore, we used the best available data for each storm.

```python
#@title Reading in SST data from NOAA
# long term monthly SST means, 1961 - 1990
sstfiledata = urllib.request.urlopen('https://downloads.psl.noaa.gov/Datasets/noaa.oisst.v2/sst.ltm.1961-1990.nc')
sstdatatowrite = sstfiledata.read()
with open('/content/data/SST_ltmean_1961_1990.nc','wb') as f:
  f.write(sstdatatowrite)
SST_ltmean_1961_1990 = xr.open_dataset('data/SST_ltmean_1961_1990.nc',engine='netcdf4',decode_times=False)

# monthly SST means, 1981 - present
sstfiledata = urllib.request.urlopen('https://downloads.psl.noaa.gov/Datasets/noaa.oisst.v2/sst.mnmean.nc')
sstdatatowrite = sstfiledata.read()
with open('/content/data/SST_mnmean_1981_present.nc','wb') as f:
  f.write(sstdatatowrite)
SST_mnmean_1981_present = xr.open_dataset('data/SST_mnmean_1981_present.nc',engine='netcdf4',decode_times=False)
```

```python
#@title Data cleaning: converting "days since 1800-01-01" time variable into months, converting date strings into datetime objects, and converting longitudes to range from -180 to 180 instead of 0 to 360
month = np.full(cost.iso_time.shape, datetime(1,1,1))
for i in np.arange(cost.dims['storm']):
  for j in np.arange(cost.dims['date_time']):
    datestring = str(cost.iso_time.values[i][j])
    if len(datestring) < 5:
      month[i][j] = np.nan
      continue
    datestring = datestring[2:-13]
    dateobject = datetime.strptime(datestring, '%Y-%m')
    month[i][j] = dateobject.date()

cost=cost.assign_coords(month = (['storm','date_time'],month))
cost.month.attrs={"Description": "Month and year as datetime object"}

# converting "days since 1800-01-01" time variable into months
time = np.full(SST_mnmean_1981_present.time.shape, datetime(1,1,1))
for i in np.arange(SST_mnmean_1981_present.dims['time']):
  time[i] = date(1800,1,1) + timedelta(SST_mnmean_1981_present.time.values[i])

SST_mnmean_1981_present = SST_mnmean_1981_present.assign_coords(time = time)
SST_ltmean_1961_1990 = SST_ltmean_1961_1990.assign_coords(time = np.arange(1,13))

# adding months variable to newer dataset, and simply assigning months jan-dec to older dataset (only has long term means)
SST_mnmean_1981_present = SST_mnmean_1981_present.assign_coords(time = time)
SST_ltmean_1961_1990 = SST_ltmean_1961_1990.assign_coords(time = np.arange(1,13))

# converting longitudes to range from -180 to 180 instead of 0 to 360, in order to match cost dataset
lons = SST_ltmean_1961_1990.lon.values
for i in np.arange(len(lons)):
  if lons[i] > 180:
      lons[i] = lons[i] - 360
SST_ltmean_1961_1990 = SST_ltmean_1961_1990.assign_coords(lon = lons).sortby('lon')

lons = SST_mnmean_1981_present.lon.values
for i in np.arange(len(lons)):
  if lons[i] > 180:
      lons[i] = lons[i] - 360
SST_mnmean_1981_present = SST_mnmean_1981_present.assign_coords(lon = lons).sortby('lon')
```

```python
display(SST_ltmean_1961_1990)
display(SST_mnmean_1981_present)
```

```python
#@title Calculating SST at each point in every storm in our merged dataset
sst = np.empty(cost.iso_time.shape)
sst[:] = np.nan
for i in np.arange(cost.dims['storm']):
  storm = cost.sel(storm=i)
  lon_lst,lat_lst = get_lon_lat(storm)
  month_lst = storm.month.values[~np.isnan(storm.lon.values)]
  for j in np.arange(len(month_lst)):
    if storm.season.values < 1961:
      sst[i][j] = np.nan
      continue  
    if storm.season.values < 1982:
      temp = SST_ltmean_1961_1990.sel({'time':month_lst[j].month,'lon':lon_lst[j],'lat':lat_lst[j]},method='nearest')
    else:
      temp = SST_mnmean_1981_present.sel({'time':month_lst[j],'lon':lon_lst[j],'lat':lat_lst[j]},method='nearest')
    closest_sst = temp.sst.values
    sst[i][j] = closest_sst
```

```python
cost=cost.assign(sst = (['storm','date_time'],sst))
cost.sst.attrs={"long_name": "sea surface temperature","Source":"NOAA Optimum Interpolation SST V2","Description": "Month and year as datetime object"}
cost
```

### 3.1 Overveiw

We chose to focus on the following variables:
- Maximum wind speed: faster storms are likely to cause more destruction once they hit land, so maximum wind speed is an important variable to consider.
- Minimum wind pressure: low wind pressure can result in weak vertical wind shear which, as mentioned above, is a favorable condition for hurricane formation.
- Power Dissipation Index (PDI): this metric is an efficient way of combining wind velocity and hurricane duration.
- Sea Surface Temperature (SST): as mentioned above, high SST is a favorable condition for hurricane formation. We considered mean SST along the hurricane track, SST at the origin, and the number of times the SST along a track was above 27 degree thresholds for a 6-hour measurement period.

Furthermore, all of these metrics were available for the majority of hurricanes in our dataset, so we did not have to shrink our dataset too much by considering them.

### 4.1 Full model

$LogEconomicLoss = 105.4244 - 0.0297 * MaxWindSpeed - 0.0938 * MinWindPressure - 0.2420 * PDI + 0.0882 * MeanSST + 0.2149 * OriginSST$


```python
# Identifying missing values in features to treat them
features.isnull().sum()
```

```python
# Clean data (delete the rows contain nan data)
features = features.dropna()
Y = features['LogEconomicLoss']

# Full model
X_full = features[['MaxWindSpeed', 'MinWindPressure', 'PDI', 'MeanSST', 'OriginSST']]
xf_train, xf_test, yf_train, yf_test = train_test_split(X_full, Y, test_size = 0.2, random_state = 4)
model = lm.LinearRegression()
model.fit(xf_train,yf_train)
yf_predict =  model.predict(xf_test)
print("Full models:")
print("LogEconomicLoss = 105.4244 - 0.0297 * MaxWindSpeed - 0.0938 * MinWindPressure - 0.2420 * PDI + 0.0882 * MeanSST + 0.2149 * OriginSST")
print("\nCoefficients:", model.coef_)
print("Intercept: ", model.intercept_)
print("\nEvaluation:")
print('r2 socre: ', r2_score(yf_test,yf_predict))
print('mean_sqrd_error: ', mean_squared_error(yf_test,yf_predict))
```

### 4.11 Residuals

```python
residual_full = yf_test-yf_predict

fig,ax=plt.subplots(1,2,figsize=(11,5))
ax[0].scatter(x=yf_predict, y=residual_full)
ax[0].axhline(y=0, linestyle='--')
ax[0].set_title('Residuals(Error Terms)')
ax[0].set_xlabel('Predicted LogEconomicLoss')
ax[0].set_ylabel('Residual')
ax[1].hist(residual_full)
ax[1].set_title('Distribution of Residuals')
```

Multicollinearity is the occurrence of high intercorrelations among two or more independent variables in a multiple regression model. Multicollinearity can lead to wider confidence intervals that produce less reliable probabilities in terms of the effect of independent variables in a model. [1]

By the correlation matrix below, the three variables -- MaxWindSpeed, MinWindPressure, and PDI -- are high correlated, so we decided to remove two of them to avoid multicollinearity.

### 5. Experimenting with new features

5.1 SST Threshold Crossing Rate

As we can see in the correlation table above the mean and the origin SST along the track aren't really contributing to the explanation of the trend in economic loss. Hence we engineered a new feature that we called "threshold crossing rate" which tells how many times (6-hour increments to be more precise) has the SST along the track of the hurricane crossed a threshold average value. We also show that the feature has better correlation with the economic loss (cost of damage)

```python
# Recompiling our features dataframe and ignoring mean SST and origin SST since as discussed above they had 
# almost no contribution in predicting economic loss and also because  the missing SST data required us to delete
# certain rows of valuable information
features = pd.DataFrame({
    'Storm Name': cost.name.values,
    'Season': cost.season.values,
    'LogEconomicLoss': np.log(cost["normalized_cost_cl"]),
    'MaxWindSpeed': np.nanmax(cost["usa_wind"].values, axis=1),
    'MinWindPressure': np.nanmin(cost["usa_pres"].values, axis=1),
    'PDI': cost['pdi'],
    'MeanSST': cost.sst.mean(axis=1),
    'OriginSST': cost.sst.sel(date_time=0)
})

```

```python
above_threshold=np.empty(cost.dims['storm'])
threshold = 27

for i in np.arange(cost.dims['storm']):
  storm = cost.sel(storm=i)
  storm_times = storm.time.values[~np.isnan(storm.time.values)]
  if np.isnan(storm.sst).all():
    above_threshold[i] = np.nan
    continue

  modified_storm_times = np.array([])
  for j in np.arange(len(storm_times)):
    curr_time = storm_times[j]
    if j == 0:
      origin_time = curr_time
      modified_storm_times = np.append(modified_storm_times, origin_time)
      continue
    delta_t = curr_time - origin_time
    if delta_t >= 0.25:
      modified_storm_times = np.append(modified_storm_times, curr_time)
      origin_time = curr_time
    else:
      modified_storm_times = np.append(modified_storm_times, np.nan)
  
  num_time_steps = np.size(modified_storm_times)
  num_above = 0
  for j in range(num_time_steps):
    curr_sst = storm.sst[j]
    if np.isnan(modified_storm_times[j]):
      continue
    elif storm.sst[j] > threshold:
      num_above = num_above + 1
  above_threshold[i] = num_above

cost=cost.assign(above_threshold = (['storm'],above_threshold))
cost.sst.attrs={"long_name": "6-hour segments above 27deg threshold","Source":"NOAA Optimum Interpolation SST V2","Description": "number of 6-hour increments of storm above 27 deg C temp threshold"}
cost
```

5.2 Nakamura Clusters (Clusters based on track information)

Here we have tried exploring whether the clusters as derived by Nakamura et al (2009) might hold any predictive power when it comes to predicting economic loss. To do this we experimented with clustering based on <br>
a.) unweighted moments<br>
b.) moments weighted by max wind speed<br>
c.) moments weighted by time difference between adjacent observations <br>

The idea was to in a way understand whether track information might play any role in predicting economic loss

```python
# Helped functions to calculate moments and plot charts

def nan_helper(y):
    return np.isnan(y), lambda z: z.nonzero()[0]

def plot_elbow(moments_data):
    """
    Plots inertia values for different number of clusters
    Input: Moments dataframe
    Output: Plots inertia (loss) loss values for different number of clusters
    """

    sum_of_squares = []
    K = range(1, 15)
    for k in K:
        km = k_means(moments_data.iloc[:,3:], n_clusters=k)
        sum_of_squares.append(km[2])
    plt.plot(K, sum_of_squares, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Sum_of_squared_distances')
    plt.title('Inertia Values against the Number of Cluster k')
  
def cluster_2D_PCA(moments_data):
    """
    Input: moments data
    Output: Visualiazes clusters by compressing moments data for hurricanes in 2D PCA 
    """

    features = ['X-centroid', 'Y-centroid', 'X_var', 'Y_var', 'XY_var']
    x = moments_data.loc[:, features].values
    # Standardizing the features
    x = StandardScaler().fit_transform(x)
    pca = PCA(n_components=2)
    principalComponents = pca.fit_transform(x)
    principalDf = pd.DataFrame(data = principalComponents
              , columns = ['principal component 1', 'principal component 2'])
    finalDf = pd.concat([principalDf, moments_data[['Clusters']]], axis = 1)

    fig = plt.figure(figsize = (8,8))
    ax = fig.add_subplot(1,1,1) 
    ax.set_xlabel('Principal Component 1', fontsize = 15)
    ax.set_ylabel('Principal Component 2', fontsize = 15)
    ax.set_title('2 component PCA', fontsize = 20)
    targets = [0,1,2,3,4,5]
    colors = ['r', 'g', 'b','y','black',"grey"]
    for target, color in zip(targets,colors):
        indicesToKeep = finalDf['Clusters'] == target
        ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
                   , finalDf.loc[indicesToKeep, 'principal component 2']
                   , c = color
                   , s = 10)
    ax.legend(targets)
    ax.grid()



def get_weighted_moments(storm,weighted_by):

  """
  Calculates weighted momments for a given storm and weights
  Inputs: Storm & Weight type indicator (unweighted vs max wind speed vs time interval)
  Outputs: Storm Name, Season, X-centroid, Y-centroid, X_var, Y_var, XY_var 
  """

  lon_lst = storm.lon.values
  lon_lst=lon_lst[~np.isnan(lon_lst)]
  lat_lst = storm.lat.values
  lat_lst=lat_lst[~np.isnan(lat_lst)]

  if len(lon_lst)<= 1: return None
  
  if weighted_by=="unweighted":
    weights=np.ones(lon_lst.shape[0])
  
  elif weighted_by=="wind":
    if np.isnan(storm.wmo_wind.values).all():
        weights=np.ones(lon_lst.shape[0])
    else:
      wind_speeds = storm.wmo_wind.values[~np.isnan(storm.lon.values)]
      nans,x = nan_helper(wind_speeds)
      wind_speeds[nans] = np.interp(x(nans),x(~nans),wind_speeds[~nans])
      weights=wind_speeds
  
  elif weighted_by=="time":
    weights=np.zeros(lon_lst.shape[0])
    for i in range(1,lon_lst.shape[0]):
      weights[i]=(storm.time.values[i]-storm.time.values[i-1])#*24*60*60
    weights[0]=np.average(weights[1:]) # time interval for the first timestamp assumed to be equal to average of time intervals for subsequent steps
  
  weights_total=sum(weights)
  
  lon_weighted, lat_weighted = sum(lon_lst*weights)/weights_total, sum(lat_lst*weights)/weights_total

  cv = np.cov([lon_lst, lat_lst],aweights=weights,ddof=0)
  name=str(storm.name.values).lower()
  season=str(int(storm.season))

  return [name,season,season+"_"+name,lon_weighted, lat_weighted, cv[0, 0], cv[1, 1], cv[0, 1]]

  
```

```python
# Extracting storms that were tracked between 1948 and 2021 as
# Nakamura et al. only studied storms starting from 1948 becuase of the accuracy of the data
storms_relevant= ds.where((ds.season >= 1948) & (ds.season <= 2022), drop=True)
```

## 6. Conclusion

**Background:** Although predictions of hurriance costs can be highly uncertain given the multitude of driving factors, analyzing historical storm data can provide a crucial first step for understanding the relationships between hurricane properties and the estimated costs. 

**Methods:** In this study, we combined the IBTrACS storm track data in the North Atlantic basin with normalized hurricane damange in the continental United States from 1950-2017 (Weinkle et al. 2018) to examine how maximum wind speed, minimum pressure, and the power dissipation index (PDI) influence the economic costs of each hurricane. We further extracted SST along the hurricane track and examined how SST at the origin or averaged along the track influence hurricane costs. To identify whether underlying patterns in hurricane tracks may impact hurricane damages, we adapted the K-means clustering techniques in Nakamura et al.(2009) and encoded the clustering results as an additional predictor. We examined the pair-wise correlation between predictor variables and removed highly correlated variables to avoid multicollinerality issues. 

**Results:** Multivariate regression between the hurricane costs and the predictor variables suggested minimum wind pressure and maximum wind speed are the most relevant features for hurricane costs followed by encoded hurricane clusters, whereas sea surface temperature has at best an indirect impact on the costs of hurricane damage. Our best model achieved an $R^{2}$ score around 0.76 when evaluated on an indepenet test dataset.

**Implications:** The study shows that predictions of hurricane costs should mostly depend on imminent meterological observations of storm intensity, which are often not directly observable before a hurricane is fully developed. Therefore, the study highlights the value in developing satellite and ground-based observational networks for hurricane early warning and designing robust methods for wind and precipitation nowcasting during an impact.
